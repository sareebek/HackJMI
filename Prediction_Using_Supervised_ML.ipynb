4-  Divide the data into "attributes" (inputs) and "labels" (outputs).

```python
X = s_data.iloc[:, :-1].values
y = s_data.iloc[:, 1].values
```

5- Splitting Training and Data sets

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y,
                            test_size=0.2, random_state=0)
```

6- Training modek on Linear regrassions with Prevuous Data

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, y_train)

print("Training complete.")
```

7- Plotting the regression line

```python

line = regressor.coef_*X+regressor.intercept_

plt.scatter(X, y)
plt.plot(X, line);
plt.show()
```

8- Testing data

```python
print(X_test)
y_pred = regressor.predict(X_test)
```

9- Comparing Actual vs Predicted

```python

df = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})
df
```

10- Prediction for 9.25 Hrs

```python
hours = 9.25
own_pred = regressor.predict([[hours]])

print("Predicted Score =",own_pred[0])
```

11- Mean Absolute Error

```python
from sklearn import metrics
print('Mean Absolute Error:',
      metrics.mean_absolute_error(y_test, y_pred))
```
